#### Linked list

```python
class Node:
    def __init__(self, data=None, next=None):
        self.data = data
        self.next = next 
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def display(self):
        if self.head is None:
            print('list is empty')
            return
        else:
            lls = ''
            itr = self.head
            while itr is not None:
                lls += str(itr.data)+'-->'
                itr = itr.next
        print(lls)
            
    #inserting at the end   
    def insert_at_end(self,data):
        if self.head is None:
            self.head = Node(data, None)
            return
        else:
            itr = self.head
            while itr.next is not None:
                itr = itr.next
            itr.next = Node(data,None)
            
        
    #creating a new linkedlist
    def insert_values(self, data_list):
        self.head = None
        for i in data_list:
            self.insert_at_end(i)
     
    #index
    def length(self):
        count = 0
        itr = self.head  # Start from the head of the list
        while itr is not None:
            count += 1
            itr = itr.next  # Move to the next node
        return count

    
    #inserting at the index
    def insert_at_index(self, index,data):   
        if index == 0:
            node = Node(data, self.head)
            self.head = node
        elif index < 0 or index > self.length():
            raise Exception('invalid index')
        else:
            count = 0
            itr = self.head#start from head
            while itr is not None:
                if count == index -1:
                    node = Node(data, itr.next)
                    itr.next = node
                itr = itr.next #move to next node and update count
                count+=1
            
    #removing at index
    def remove_at(self, index):
        if index == 0:
            self.head = self.head.next
        elif index < 0 or index > self.length():
            raise Exception('invalid index')
        else:
            count = 0
            itr = self.head
            while itr is not None:
                if count == index - 1:
                    itr.next = itr.next.next #next node should be next of the next node
                itr = itr.next
                count +=1
                
    #inserting after
    def insert_after_value(self,data_after,data_to_insert):
        itr = self.head
        while itr is not None:
            if itr.data == data_after:
                node = Node(data_to_insert, itr.next)
                itr.next = node
                break
            itr = itr.next
            
    #remove by value
    def remove_by_value(self,data):
        if self.head == data:
            self.head = self.head.next
            
        itr = self.head
        while itr.next is not None:
            if itr.next.data == data:
                itr.next = itr.next.next
            itr = itr.next
        

            
ll = LinkedList()
ll.insert_values(['pech', 'peech','peach','preach'])
ll.display()
ll.insert_at_index(0,'man')
ll.display()
ll.insert_at_index(3,'pork')
ll.display()
print('length is :',ll.length())
ll.remove_at(4)
ll.display()
print('length is :',ll.length())
ll.insert_after_value('pork', 'perch')
ll.display()
ll.remove_by_value('pech')
ll.display()
```

#### Double linked list

```python
class Node:
    def __init__(self, data=None, next=None, previous=None):
        self.data = data
        self.next = next
        self.previous = previous

class LinkedList:
    def __init__(self):
        self.head = None
    
    def print_forward(self):
        if self.head is None:
            print('list is empty')
            return
        itr = self.head
        lls = ''
        while itr is not None:
            lls += str(itr.data) + '-->'
            itr = itr.next
        print(lls)
        
    def print_backward(self):
        if self.head is None:
            print('List is empty')
            return
        # Find the last node
        itr = self.head
        while itr.next is not None:
            itr = itr.next
        # Traverse backwards and print the nodes
        lls = ''
        while itr is not None:
            lls += str(itr.data) + ' --> '
            itr = itr.previous
        print(lls)
        
    def insert_at_end(self, data):
        if self.head is None:
            self.head = Node(data, None)
            return
        itr = self.head
        while itr.next is not None:
            itr = itr.next
        itr.next = Node(data, None, itr)  # Set the previous node of the new node
                                           # to be the current last node
        
    def insert(self, data_list):
        for i in data_list:
            self.insert_at_end(i)

ll = LinkedList()
ll.insert(['jj', 'll', 'kk', 'hh'])
ll.print_forward()
ll.print_backward()
```
